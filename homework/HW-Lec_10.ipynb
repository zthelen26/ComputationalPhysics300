```python
#import mods
import numpy as np
import matplotlib.pyplot as plt
from scipy import pi
from scipy.fftpack import fft
```

```python

rate = 1024     #samples/sec
N = (2-0)*rate  #array size

t = np.linspace(0,2,N) #time

fr1 = 60   #frequency 
mag1 = 25  #magnitude
fr2 = 270
mag2 = 2 

wave1 = mag1*np.sin(2*pi*fr1*t)  #wavefunction
wave2 = mag2*np.sin(2*pi*fr2*t)

noise = np.random.normal(0,3,N)  #noisefunction

data_t = wave1 + wave2 + noise

plt.figure(figsize=[12, 8]) #blow up image
plt.plot (t[0:100], data_t [0:100])
plt.xlabel ('Time')
plt.ylabel ('A')
plt.show ()
```

```python
frequency = np.linspace(0.0, 512, int(N/2)) #new frequency w/ range .5 of rate (due to Nyquist-shannon theorem)

data_fr = fft(data_t)                       #fourier transform of data
y = 2/N*np.abs(data_fr[0:np.int(N/2)])      #y axis for grpahing

plt.figure(figsize=[12, 8]) #blow up image
plt.plot(frequency, y)
plt.title('Frequency')
plt.xlabel('Frequency in Hz')
plt.ylabel('A')
plt.show()
```

```python
#I used fourier transform applied in vibration analysis. 
#In the first graph you can see the measurment over time is hard to read with conslusive variables for the frequency and
#   magnitude in wave form created from the "vibration". 
#Using the fft method through scipy we have an automated grid that has defined spikes that can be read 
#   as the original variables of frequency and proper magnitude.
```
